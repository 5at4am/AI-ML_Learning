```python
cust = pd.DataFrame({
    "cust_id":[1,2,3,4],
    "name":["a","b","c","d"]
})

order = pd.DataFrame({
    "order_id":[101,102,103,104],
    "cust_id":[5,1,2,4],
    "amount":[100,250,130,240]
})
```

```python
cust
```

```python
order
```

```python
pd.merge(cust,order,on="cust_id",how="inner") 
```

```python
pd.merge(cust,order,on="cust_id",how="left") 
```

```python
pd.merge(cust,order,on="cust_id",how="right") 
```

```python
pd.merge(cust,order,on="cust_id",how="outer") 
```
