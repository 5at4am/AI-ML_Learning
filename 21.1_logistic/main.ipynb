```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
X,y= make_classification(n_samples = 1000,n_features=10,n_classes= 2, random_state= 15 )
```

```python
X = pd.DataFrame(X)
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size =0.30, random_state=42)
X_train.shape,X_test.shape
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(X_train,Y_train)
```

```python
y_pred = logistic.predict(X_test)

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score= accuracy_score(Y_test,y_pred)

con=confusion_matrix(Y_test,y_pred)

clas=classification_report(Y_test,y_pred)

print(score)
print(con)
print(clas)
```

# Hyper Parameter tuning


## Grid CV

```python
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_value=[100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']


param = dict(penalty = penalty,C=c_value,solver = solver)
param
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold()

grid = GridSearchCV(estimator=model, param_grid=param,scoring='accuracy',cv=cv, n_jobs=-1)

grid
```

```python
grid.fit(X_train,Y_train)
```

```python
y_pred = grid.predict(X_test)

print(grid.best_params_)
```

```python
grid.best_score_
```

```python
score= accuracy_score(Y_test,y_pred)

con=confusion_matrix(Y_test,y_pred)

clas=classification_report(Y_test,y_pred)

print(score)
print(con)
print( clas)
```

## Randomized search cv


```python

from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv = RandomizedSearchCV(estimator = model, param_distributions = param,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,Y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred = randomcv.predict(X_test)
```

```python
score= accuracy_score(Y_test,y_pred)

con=confusion_matrix(Y_test,y_pred)

clas=classification_report(Y_test,y_pred)

print(score)
print(con)
print( clas)
```
