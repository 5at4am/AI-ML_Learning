### cleaning and structuring the data
 - data inconsistency - mixed data
 - missing data 
 - duplicated data





```python
# clean the data

def clean(data):
    # map the string to number value

    text2num = {"one":1,"two":2,"three":3,"four":4,"five":5,"six":6,"seven":7,"eight":8,"nine":9,"ten":10}

    clean_data = []
    unique_data = set()     # for the unique element

    # loop for all the assignment
    for user in data:

        # conert the data into a string for better implementation
        # strip with turn "_a_" to "a" and in lower case to map the value 
        raw_rating = str(user["rating"]).strip().lower()    

        # transform the string data in numric value 
        if(raw_rating in text2num):

            raw_rating = text2num[raw_rating]
        user["rating"]= raw_rating

        
        
        # user missing value 

        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None


        # handle duplicate data

        if(user["name"].strip() in unique_data):
            continue
        unique_data.add(user["name"])
        clean_data.append(user)

    return clean_data       # return the data from the Function
```

```python
data = clean(data)      # save the data 
data
```

## Assignment 2
- whats the avg rating?
- whats % of user gave poor rating(<3)?

```python
### get meaningfull insight

def insight(data):

    # avg rating
    total = 0
    for user in data:
        total += float(user["rating"])

    print(total/len(data))

    # % of user with poor rating
    poor = 0
    for user in data:

        if( float(user["rating"]) < 3):
            poor +=1
    print(poor/len(data)*100)
    
```

```python
insight(data)
```

### Assignment 3
- builda a recommendation feature (rule base, not ML)

    - if user rating >= 4 recommend same brand products.
    - if user rating <4, recommend different brand products

```python
# build the recommendation feature

def get_recommendation(data):
    recommendation = []

    for user in data:
        curr={}
        curr["name"] = user["name"]
        if(float(user["rating"])>=4):
            curr["brand"] = "Apple"
        else:
            curr["brand"] = "Samsung"

        recommendation.append(curr)
    return recommendation
         
```
