```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr, 25)
Q2 = np.percentile(arr, 50)
Q3 = np.percentile(arr, 75)
```

```python
print(Q1,Q2,Q3)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
UF = Q3+(1.5*IQR)
print(UF)
```

```python
LF = Q1-(1.5*IQR)
print(LF)
```

```python
l= []
for i in arr:
    if i>LF and i<UF:
        l.append(i)
l = np.array(l)
print(l)
```

```python
sns.boxplot(x = arr)
```

```python
sns.boxplot(l)
```
