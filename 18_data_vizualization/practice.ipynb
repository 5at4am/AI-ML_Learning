```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#weekly temprature 
days = ["mon","tue","wed","thur","fri","sat","sun"]

cities = ["New York","London","Delhi","Tokyo"]

temp = [
    [22,23,21,24,26,25,27],
    [21,26,23,22,22,24,20],
    [25,25,24,23,20,21,21],
    [20,22,22,21,21,22,24]

]
color = ["r","g","b","y"]
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,10))

cityn= 0
for i in range(2):  # rows 
    for j in range(2):  # column
        ax[i,j].plot(days, temp[cityn], marker = "o", color = color[cityn])
        ax[i,j].set_title(cities[cityn])
        ax[i,j].grid(True)
        cityn += 1



fig.suptitle("temprature")
fig.tight_layout()
```

```python
fig, ax = plt.subplots()

for i in range(4):
    ax.plot(days, temp[i], marker = "o", label = cities[i])

ax.set_title("temprature")
ax.grid(True)
ax.legend()
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,10))
axes = ax.flatten()

for i,ax in enumerate(axes):
    ax.plot(days,temp[i],marker="o",color=color[i])
    ax.set_title(cities[i])
    ax.grid(True)
```
