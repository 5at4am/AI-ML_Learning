### Linkdlist

```python
class Node:     # creating a node
    ## initializing the node
    def __init__ (self, data, next = None):
        self.data = data                # data
        self.next = next                # next

# creating a linked list        
class LinkedList:   
    ## initializing the linked list

    def __init__(self, head = None):
        self.head = head                # head

    ## insertion at the beginning
    def in_beg(self, data):
        # creation of new node
        new_node = Node(data)           # creating a new node
        # insertion at the beginning
        new_node.next = self.head       # next of new node points to head
        self.head = new_node            # head points to new node

    ## insertion at the end
    def at_end(self,data):  
        new_node = Node(data)       # creating a new node
        temp = self.head            # temp points to head
        while temp.next:            # while temp.next is not null
            temp = temp.next        # temp points to next
        temp.next = new_node        # next of temp points to new node

    def display(self):
        temp = self.head
        while temp:
            print(str(temp.data) + '->', end = ' ') # print(temp.data)
            temp = temp.next

list = LinkedList()                 # creating a linked list
list.in_beg(20)                     # insertion at the beginning
list.in_beg(30)
list.in_beg(10)
list.in_beg(40)
list.at_end(50)                     # insertion at the end
list.display()                      # display the linked list



        
    
```

### doubli linkedlist

```python
## creating a doubli linkedlist

class Node:
    def __init__(self, data, next = None, prev = None):
        self.data = data
        self.next = next
        self.prev = prev

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def in_beg(self, data):
        new_node = Node(data)
        new_node.next = self.head
        if self.head:
            self.head.prev = new_node
        self.head = new_node
        
    def in_end(self, data):
        new_node = Node(data)
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
        
    def display(self):
        temp = self.head
        while temp:
            print(str(temp.data) + '->', end = ' ')
            temp = temp.next
            
list = DoublyLinkedList()
list.in_beg(10)
list.in_beg(20)
list.in_end(30)
list.display()
```
