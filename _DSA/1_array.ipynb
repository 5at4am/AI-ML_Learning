## Two Sum 
##### Finding a pair of items in a list that sum up to a specific target. Commonly used in e-commerce (finding two products that fit a budget) or financial analysis.




### Steps
```
1. Initialize an empty dictionary seen.
2. Iterate through the array with index i and value num.
3. Calculate complement = target - num.
4. Check if complement exists in seen.
    - If yes, return the index of complement and current index i.
    - If no, store num in seen with its index i.
```

### Pseudocode 
```
function twoSum(nums, target):
    create empty map seen
    for i from 0 to length of nums - 1:
        complement = target - nums[i]
        if complement is in seen:
            return [seen[complement], i]
        else:
            seen[nums[i]] = i
    return []
```

````python
def TwoSum(arr,target):
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i]+arr[j]==target:
                return i,j
````

````python
arr = array('i',[2,5,3,9,11])
targer = 13
print(f"index at :  {TwoSum(arr,targer)}")
````

##  Maximum Subarray

#### Determining the contiguous subarray with the largest sum. Used in stock analysis to find the best period to hold a stock or in image processing for region detection.



### Logic:

1. Initialize max_so_far to the first element (or negative infinity).
2. Initialize current_max to 0.
3. Loop through each number x in the array.
4. Update current_max by adding x.
5. If current_max is greater than max_so_far, update max_so_far.
6. If current_max becomes negative, reset it to 0.

### Pseudocode
```
function maxSubArray(nums):
    max_so_far = -infinity
    current_max = 0
    for x in nums:
        current_max = current_max + x
        if max_so_far < current_max:
            max_so_far = current_max
        if current_max < 0:
            current_max = 0
    return max_so_far
```

````python
def MaxSubarray(arr):
    max = float('-inf')
    curr = 0
    for i in range (len(arr)):
        curr += arr[i]
        if(curr>max):
            max = curr
        if(curr<0):
            curr = 0
    return max
````

````python
arr = array('i',[-2, 1, -3, 4, -1, 2, 1, -5, 4])
print(f" your max sum is : {MaxSubarray(arr)} at the index {TwoSum(arr,MaxSubarray(arr))}")
````
