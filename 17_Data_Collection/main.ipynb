```python
# webscrapping 


URL = "https://www.scrapethissite.com/pages/simple/"

res = requests.get(URL)
```

```python
# status code
print(res.status_code)
```

```python
if res.status_code == 200:
    print("Success")
    print(res.headers)
else:
    print("Error")
 
with open("scrape/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
    f.close()
```

```python
try:
    r = requests.get("https://www.scrapethissite.com/pages/simple/", timeout=5)
    r.raise_for_status()
except requests.exceptions.RequestException as e:
    print("error",e)
```

```python
from bs4 import BeautifulSoup as bs

with open("scrape/data1.html","r",encoding="utf-8") as f:
    html_content = f.read()

soup = bs(html_content,"lxml")
```

```python
soup.find("h3")
soup.h3
```

```python
all_h3 = soup.find_all("h3")

type(all_h3)
```

```python
for h3 in all_h3:
    print(h3.text)
```

```python
for h3 in all_h3:
    print(h3.text.strip())
```

```python
for h3 in all_h3:
    print(h3.get_text(strip=True))
    print(h3.find_parent("div")["class"])
```
