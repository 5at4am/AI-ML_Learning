```python
# import ols model
import statsmodels.formula.api as sm
```

```python
model = sm.ols('y~x',data = df).fit()
```

```python
# find the model summery
model.summary()
```

```python
model2 = sm.ols('y~np.log(x)',data = df).fit()
model2.summary()
```

```python
model3 = sm.ols('np.log(y)~x',data = df).fit()
model3.summary()
```

```python
model4 = sm.ols('np.log(y)~np.log(x)',data = df).fit()
```

```python
model4.summary()
```

```python
model5 = sm.ols('np.log(y)~x+I(x*x) ',data = df).fit()   
```
